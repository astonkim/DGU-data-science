```python
# 축약된 두 문자 헤더 교체하기
# data01.csv 파일의 축약된 두 문자 헤더에 대한 설명을 추가한 csv 데이터 파일을 만드시오.
# data01.csv : 축약된 두 문자 헤더를 포함하는 csv 데이터 파일
# header01.csv : 데이터 파일의 거의 대부분 헤더 설명이 있는 csv 데이터 파일
# header02.csv : header01.csv에는 없는 몇몇 헤더 설명이 있는 csv 데이터 파일
# 이 파일들의 encoding은 default로 한다.

from csv import reader
data01 = reader(open('data01.csv', 'rt', encoding='cp949'))
header01 = reader(open('header01.csv', 'rt', encoding='cp949'))
header02 = reader(open('header02.csv', 'rt', encoding='cp949'))


data01_header = next(data01)

header01_header = [d for d in header01]
header01_header01 = [d[0] for d in header01_header]

header02_header = [h for h in header02]
header02_header02 = [h[0] for h in header02_header]

print(len(data01_header))
print(len(header01_header01))
print(len(header02_header02))

same_header = []
for d in data01_header:
    if d in header01_header01:
        same_header.append(d)
print(same_header)
print(len(same_header))

# 공통되지 않은 data01의 header를 출력하고, 이를 mismatch 변수에 저장하시오.
mismatch = []
for d in data01_header:
    if d not in header01_header01:
        mismatch.append(d)
print(mismatch)
print(len(mismatch))

# mismatch에 들어가 있는 header 중 header02에 있는 header를 찾아서 출력하시오.
# 이를 header02_match 변수에 저장하시오.
header02_match = []
for m in mismatch:
    if m in header02_header02:
        header02_match.append(m)
print(header02_match)
print(len(header02_match))

# 최종적으로 전처리된 데이터 파일의 header 개수는?
# 이를 header_final 변수에 저장하시오.
header_final = same_header + header02_match
print(len(header_final))
```
