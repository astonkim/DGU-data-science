```python
df = pd.read_csv('data-text-100.csv', encoding='utf-8', header=None)
# 뒤에서 20행에 대한 'Indicator', 'Year", 'Country', 'Sex' 필드만 추출하여 데이터프레임을 만들고 출력하시오.
df2 = df.iloc[-20:, [0, 2, 5, 6]]
print(df2)
```

```python
# 2번째 행의 'Year'필드를 1990년에서 2002년으로 수정하고 SQLite 데이터베이스에 저장하시오
df2.iloc[1, 1] = 2002
print(df2)
```

```python
import sqlite3
con = sqlite3.connect('test.db')
df2.to_sql('data-text-100', con, if_exists='append', index=False)
con.close()
```
