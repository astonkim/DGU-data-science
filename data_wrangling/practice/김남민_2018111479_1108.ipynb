```python
import seaborn as sns
df = sns.load_dataset('titanic')
df
```

```python
# 타이타닉호에 승선한 도시를 나타내는 'embark_town'열에서 누락 데이터가 있는지 평가하여 있으면 그 개수를 출력하시오.
df[df['embark_town'].isnull()]
```

```python
df['embark_town'].isnull().sum()
```

```python
# 누락 데이터 중 앞에서부터 발견한 첫 번째 누락 데이터에 대해 그 앞뒤 2개씩 총 5개의 승선 도시를 출력하시오.
df['embark_town'][61-2:61+3]
```

```python
# 승선 도시 중 가장 많이 승선한 도시를 찾아서 출력하시오.
df['embark_town'].value_counts().head(1)
```

```python
# 승선 도시 최빈값으로 누락 데이터를 대체하고, 앞의 5개 데이터를 다시 출력하시오.
df['embark_town'].fillna(df['embark_town'].value_counts().head(1).index[0], inplace=True)
```
